```python
import os
import mne
import json
import pandas as pd
from pathlib import Path
import re
import matplotlib.pyplot as plt
import numpy as np

import warnings
```

```python
# === CONFIGURATION ===
subject = 'NDARAC904DMU'  # change this to any subject ID
data_dir = Path('/mount/sub/cmi_bids_R1/eeg')
```

```python
# === FUNCTION TO BUILD FILE PATHS ===
def get_file(task, ext, run=None):
    name = f"sub-{subject}_task-{task}"
    if run:
        name += f"_run-{run}"
    return data_dir / f"{name}_{ext}"
```

```python
# === FIND ALL EEG .set FILES FOR SUBJECT ===
eeg_files = sorted(data_dir.glob(f"sub-{subject}_task-*_eeg.set"))

# Extract task and optional run info
task_info = []
pattern = re.compile(rf"sub-{subject}_task-(?P<task>.+?)(?:_run-(?P<run>\d+))?_eeg.set")

for file in eeg_files:
    match = pattern.match(file.name)
    if match:
        task_info.append({
            "task": match.group("task"),
            "run": match.group("run") 
        })

print("Found task/run pairs:")
for info in task_info:
    print(f"- Task: {info['task']}  Run: {info['run'] or 'N/A'}")
```

```python
# === VISUALIZATION LOOP ===

for info in task_info:
    task = info['task']
    run = info['run']

    print(f"\n--- Loading task: {task}" + (f"  Run: {run}" if run else "") + " ---")

    # Load EEG data
    eeg_path = get_file(task, 'eeg.set', run=run)
    raw = mne.io.read_raw_eeglab(eeg_path, preload=True,montage_units='cm')

    # Set 128-channel GSN 200 v.2.1
    montage = mne.channels.make_standard_montage('GSN-HydroCel-128')
    raw.drop_channels(['Cz'])
    raw.set_montage(montage, match_case=False)
    mne.viz.plot_sensors(raw.info, show_names=True)
    raw.plot_sensors(show_names=True)
    raw.plot_sensors('3d')

    # Load and print metadata
    json_path = get_file(task, 'eeg.json', run=run)
    if json_path.exists():
        with open(json_path) as f:
            eeg_metadata = json.load(f)
        print("Metadata:\n", json.dumps(eeg_metadata, indent=4))

    # Load and print events
    event_path = get_file(task, 'events.tsv', run=run)
    if event_path.exists():
        events = pd.read_csv(event_path, sep='\t')
        print("Events:\n", events.head())

    # Load and print channels
    channels_path = get_file(task, 'channels.tsv', run=run)
    if channels_path.exists():
        channels = pd.read_csv(channels_path, sep='\t')
        print("Channels:\n", channels.head())

    # Load and print electrodes (if available)
    electrodes_path = get_file(task, 'electrodes.tsv', run=run)
    if electrodes_path.exists():
        electrodes = pd.read_csv(electrodes_path, sep='\t')
        print("Electrodes:\n", electrodes.head())

    # Plot EEG (time domain)
    plot_title = f"{subject} - {task}" + (f" (Run {run})" if run else "")
    raw.plot(n_channels=10, title=plot_title, scalings='auto', show=True, block=True)

    # Plot Power Spectral Density (frequency domain)
    psd = raw.compute_psd(fmax=60)
    psd.plot(average=True, spatial_colors=False, dB=True, show=True)
```
